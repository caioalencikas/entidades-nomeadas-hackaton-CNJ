```python
import re
import itertools
    
def find_pattern_mentions(regex_pattern, input_str):
    mentions = []
    matches = re.finditer(regex_pattern, input_str, re.MULTILINE)

    for matchNum, match in enumerate(matches, start=1):
#         print ("Match {matchNum} was found at {start}-{end}: {match}".format(matchNum = matchNum, start = match.start(), end = match.end(), match = match.group()))
        mentions.append(match.group())
        
    return mentions

def parse_legal_entities(_input_text):

    resultados = {"artigo":[],"lei":[],"dispositivo":[], "peticaoInicial":[]}
    
    
    for categoria_de_regra in regex_patterns.keys():
#         print("procurar por padrões de : ", categoria_de_regra)
        for regra in regex_patterns[categoria_de_regra]:
            try:
                found_mentions = find_pattern_mentions(regra, _input_text)
                if(len(found_mentions) > 0):
                    resultados[categoria_de_regra].append(found_mentions)
            except:
#                 print("regra inválida =>", regra)
                pass



    def flatten_list(_list):
        return list(itertools.chain(*_list))

    for _ in ["artigo","lei","dispositivo", "peticaoInicial"]:
        resultados[_] = flatten_list(resultados[_])

    return resultados    
```

```python
def criar_df_artigo_lei(resultado):
    # criando df com linha para cada conjunto de artigo, lei e dispositivo_legal
    lista_dicts = []
    for index, row in resultado.iterrows():
        dict_codigos = {"artigo": "", "dispositivo_legal": ""}
        for artigo in row['Artigo']:
            #print(artigo)
            artigo_lista = re.findall(r"\d+", artigo)
            if len(artigo_lista) == 0:
                continue
            #print(artigo_lista)
            artigo_num = artigo_lista[0]
            dict_codigos["artigo"] = artigo_num
            if (len(row['Lei'])) > 0:
                for lei in row["Lei"]:
                    dict_codigos["dispositivo_legal"] = lei
                    lista_dicts.append(dict_codigos)
                    #print(lei)
                    #print(dict_codigos)
                    dict_codigos = {"artigo": artigo_num, "dispositivo_legal": ""}
            if (len(row['Dispositivo'])) > 0:
                for dispositivo in row["Dispositivo"]:
                    dict_codigos["dispositivo_legal"] = dispositivo
                    lista_dicts.append(dict_codigos)
                    #print(dispositivo)
                    #print(dict_codigos)
                    dict_codigos = {"artigo": artigo_num, "dispositivo_legal": ""}

    df_peticao = pd.DataFrame({"artigo": str(""), "dispositivo_legal": ""}, index=[0])
    for dic in lista_dicts:
        df_prov = pd.DataFrame(dic, index=[0])
        df_peticao = df_peticao.append(df_prov)
    nan_value = float("NaN")
    df_peticao.replace("", nan_value, inplace=True)
    df_peticao.dropna(inplace=True)
    #print(df_peticao)
    return df_peticao
```

```python
def depara_codigo_assunto(df_peticao):
    #cruzando df do resultado da petição com depara de assuntos para trazer o código do assunto
    lista_cod_assuntos_df =[]
    lista_cod_pai_df = []
    column_found = ''
    for index, row in df_peticao.iterrows():
        lista_index = []
        lista_cod_assuntos = []
        lista_cod_assuntos_linha =[]
        lista_cod_pai_linha = []
        #print(f"ROW ------> {row}")
        artigo = row["artigo"]
        regex_artigo = rf"{artigo}"
        codigo = row[1]
       # print(f"ARTIGO ------>{artigo}")
       # print(f"CODIGO ------>{codigo}")
        for column in nomes_dispositivos.columns:
            for nome_lei in nomes_dispositivos[column]:
                if codigo.upper() == nome_lei:
                    column_found = column
                    #print("COLUMN FOUND")
        if column_found == '':
            lista_cod_assuntos_df.append(lista_cod_assuntos_linha)
            lista_cod_pai_df.append(lista_cod_pai_linha)
            continue
        for nome in nomes_dispositivos[column_found]:
            regex_codigo = rf"(?i){nome}"
            for index, row in assuntos_df.iterrows():
                #print(row['dispositivo_legal'])
                #if codigo.upper() == row['dispositivo_legal'].upper():
                if len(re.findall(regex_codigo, row['dispositivo_legal'])) > 0:
                    lista_index.append(index)   
        #print(f'LISTA ----------> {lista_index}')   
        if len(lista_index) > 0:
            for index in lista_index:
                if len(re.findall(regex_artigo, assuntos_df.iloc[index]['artigo'])) > 0:
                    #print(index)
                    #print(assuntos_df.iloc[index])
                    lista_cod_assuntos_linha.append(assuntos_df.iloc[index]['cod_assunto'])
                    lista_cod_pai_linha.append(assuntos_df.iloc[index]['cod_assunto'])
                    #print("Append OK")
        lista_cod_assuntos_df.append(lista_cod_assuntos_linha)
        lista_cod_pai_df.append(lista_cod_pai_linha)

        #print("Append segundo")

    # limpando a vírgula dos assuntos
    lista_cod_assuntos_df_limpa = []
    for lista in lista_cod_assuntos_df:
        lista_limpa = []
        for item in lista:
            lista_limpa.append(item.replace(',', ''))
        lista_cod_assuntos_df_limpa.append(lista_limpa)
    lista_cod_pai_df_limpa = []
    for lista in lista_cod_pai_df:
        lista_limpa = []
        for item in lista:
            lista_limpa.append(item.replace(',', ''))
        lista_cod_pai_df_limpa.append(lista_limpa)

    # inserindo no df o código do assunto
    df_peticao['cod_assunto'] = lista_cod_assuntos_df_limpa
    df_peticao['cod_item_pai'] = lista_cod_pai_df_limpa
    return df_peticao
```

```python
def gerar_recomendacao(df_peticao):
    #gerando lista de recomendações
    lista_assuntos = []
    for assunto_lista in df_peticao['cod_assunto']:
        for assunto in assunto_lista:
            #print(assunto)
            lista_assuntos.append(assunto)
    total_recomendacoes = 5
    lista_recomendacoes = []
    for i in range(total_recomendacoes):
        if i == 0:
            result = max(lista_assuntos,key=lista_assuntos.count)
            #print(result)
            lista_recomendacoes.append(result)
        else:
            while result in lista_assuntos:
                lista_assuntos.remove(result)
            lista_assuntos
            result = max(lista_assuntos,key=lista_assuntos.count)
            #print(result)
            lista_recomendacoes.append(result)
    sgt_assuntos = pd.read_csv('dados/sgt_assuntos.csv', sep=';')
    #sgt_assuntos.astype(str)
    lista_recomendacoes_descricao = []
    for item in lista_recomendacoes:
        #print(item)
        if len(sgt_assuntos[sgt_assuntos['codigo'] == int(item)]['descricao']) > 0:
            #print(sgt_assuntos[sgt_assuntos['codigo'] == int(item)]['descricao'].item())
            lista_recomendacoes_descricao.append(sgt_assuntos[sgt_assuntos['codigo'] == int(item)]['descricao'].item())
    return lista_recomendacoes_descricao
```

```python
lista_recomendacoes = []
df_recomendacoes = pd.DataFrame(columns=['arquivo', 'recomendacoes'])
#files = [filename3]

for filepath in files:
    print(filepath)
    filename = filepath+'.doc'
    file = open(filename, 'r', encoding="ISO-8859-1")
    doc = ''
    for line in file:
        doc += line.rstrip('\n')
    file.close()
    document = nlp(doc)

    # check petição inicial
    c = 0
    peticoes = set()
    for sent in document.sents:
        sentence = sent.text.strip()
        legal_entities = parse_legal_entities(sentence)
        if c >= 5:
            break
        if(len(legal_entities["artigo"]) > 0 or len(legal_entities["lei"]) > 0 
           or len(legal_entities["dispositivo"]) > 0 or len(legal_entities["peticaoInicial"]) > 0):
            for peticao in legal_entities["peticaoInicial"]:
                peticoes.add(peticao.lower())
            c += 1

    endereco = 0
    estado_civil = 0
    rg = 0
    cpf = 0
    for peticao in peticoes:
        for pattern in regex_endereco:
            for found in re.findall(pattern, peticao):
                if found == peticao:
                    endereco = 1
        for pattern in regex_estado_civil:
            for found in re.findall(pattern, peticao):
                if found == peticao:
                    estado_civil = 1
        for pattern in regex_rg:
            for found in re.findall(pattern, peticao):
                if found == peticao:
                    rg = 1
        for pattern in regex_cpf:
            for found in re.findall(pattern, peticao):
                if found == peticao:
                    cpf = 1

    requisitos_encontrados = endereco + estado_civil + rg + cpf
    
    # extração artigos, leis e dispositivos
    frases = []
    artigos = []
    leis = []
    dispositivos = []
    peticao_inicial = []

    if requisitos_encontrados >= 3:
        for sent in document.sents:
            sentence = sent.text.strip()
            #print("FRASE------->",sentence) 
            legal_entities = parse_legal_entities(sentence)
            if len(legal_entities["artigo"]) > 0 and (len(legal_entities["lei"]) > 0 
               or len(legal_entities["dispositivo"]) > 0):
                #print("FRASE------->",sentence)
                #print("ENTIDADES--->",legal_entities)
                #print()
                frases.append(sentence)
                artigos.append(set(legal_entities["artigo"]))
                leis.append(set(legal_entities["lei"]))
                dispositivos.append(set(legal_entities["dispositivo"]))
                peticao_inicial.append(set(legal_entities["peticaoInicial"]))

                resultado = pd.DataFrame(data={'Frase': frases, 'Artigo': artigos
                                               , 'Lei': leis
                                               , 'Dispositivo': dispositivos
                                               , 'Petição Inicial': peticao_inicial})
    else:
        resultado = pd.DataFrame(data={'Frase': '', 'Artigo': [], 'Lei': []
                                 , 'Dispositivo': [], 'Petição Inicial': []})

    if len(resultado) > 0:
        resultado.to_csv(f'{filepath}.csv', encoding="ISO-8859-1")
        df_peticao = criar_df_artigo_lei(resultado)
        df_peticao = depara_codigo_assunto(df_peticao)
        lista_recomendacoes_descricao = gerar_recomendacao(df_peticao)

        file_name = filepath.strip('dados')
        recomendacoes = {file_name: lista_recomendacoes_descricao}
        df_recomendacao = pd.DataFrame(list(recomendacoes.items()),columns = ['arquivo','recomendacoes'])
        df_recomendacoes = df_recomendacoes.append(df_recomendacao)
df_recomendacoes.to_csv('recomedacoes.csv', encoding="ISO-8859-1")
```
